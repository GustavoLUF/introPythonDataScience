```python
df.loc[df['promocao'] == 'Primeira', 'promocao'] = 1
df.loc[df['promocao'] == 'Segunda', 'promocao'] = 2
df.loc[df['promocao'] == 'Terceira', 'promocao'] = 3

df['promocao'] = pd.to_numeric(df['promocao'])
```

```python
df.loc[df['tamanho'] == 'Small', 'tamanho'] = 1
df.loc[df['tamanho'] == 'Medium', 'tamanho'] = 2
df.loc[df['tamanho'] == 'Large', 'tamanho'] = 3

df['tamanho'] = pd.to_numeric(df['promocao'])
```

```python
le_semana = LabelEncoder()

df['semana'] = le_semana.fit_transform(df['semana'])

le_semana.classes_
```
