Função loc para busca de valores dentro de coluna ou do dataframe

```python
df.loc[df['idade'] == 37]
```

```python
df.loc[df['idade'] == 37].shape
```

```python
df.loc[df['estado_civil'] == 'divorced'].loc[df['idade'] <27]
```

Melhor pratica de realizar o filto com a função loc é a seguinte:

```python
df.loc[(df['estado_civil'] == 'divorced') & (df['idade'] < 27)]
```

```python
df.loc[df['estado_civil'] == 'divorced', 'estado_civil'] = 'Casado'
```

```python
df.loc[df['estado_civil'] == 'Casado']
```

```python
df.loc[df['estado_civil'] == 'single', 'saldo_bancario'] = 10000
```

```python
df.loc[df['estado_civil'] == 'single'][['idade', 'emprego']]
```

```python
df.to_csv('dados/banco_alterado.csv', index=False)
```
